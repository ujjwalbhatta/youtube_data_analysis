```python
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv('netflix_vids.csv')
```

```python
df
```

```python
df.drop(df.columns[df.columns.str.contains('Unnamed')],axis = 1, inplace = True) 
```

```python
df.head(50)
```

```python
def search_movie(movie_name):
    return(df['Title'] == movie_name)
    
```

```python
filt = search_movie('Breaking Bad') 
```

```python
desired_movie = df.loc[filt]
```

```python
desired_movie
```

```python
filt1 = (desired_movie['Views'] == desired_movie['Views'].max())
desired_movie = desired_movie.loc[filt1]
desired_movie
```

```python
desired_movie['Title']
```

```python
plt.style.use("fivethirtyeight")

a = int(desired_movie['Likes'])
b = int(desired_movie['Dislikes'])
print(a,b)

slices = [a,b]
labels = ['Likes','Dislikes']

explode = [0,0.1]    #To emphasize a piece ; Here Python is emphasized


plt.pie(slices ,labels=labels, explode=explode,shadow=True,autopct='%1.1f%%', wedgeprops={'edgecolor':'black'})

c = desired_movie.get(['Title'])
d =  str(c).split()[2:]
title=' '.join(d)

plt.title(str(title))
plt.tight_layout()
plt.show()
```

```python
plt.style.use('seaborn')

data=pd.read_csv('netflix_vids.csv')
view_count = data['Views']
likes = data['Likes']
dislike = data['Dislikes']

ratio= (100-((dislike/likes)*100))
# print(ratio)
plt.scatter(view_count, likes, c=ratio, cmap='summer', linewidth=1, alpha=0.9)

cbar =plt.colorbar()
cbar.set_label('Like/Dislike ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Netflix Movies')
plt.xlabel('View Count')
plt.ylabel('Likes Count')

plt.tight_layout()
plt.show()
```
